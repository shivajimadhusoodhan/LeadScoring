```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("/Users/Sharath/Downloads/Leads.csv")
```

```python
# Check the head of the dataset
df.head()
```

```python
# Checking for data types
df.info()

```

```python
df.describe()
```

```python
# checking for null values 
(df.isnull().sum()/df.shape[0])*100
```

```python
#Removing columns with null value percentage more than 35%
for i in df.columns:
    if (df[i].isnull().mean())*100>35:
        df.drop(i,axis=1,inplace=True)
```

```python
# checking for null values /missing values again
(df.isnull().sum()/df.shape[0])*100
```

```python
#Replacing Select values with NUll which can be imputed with value 'UnKnown'
df.replace('Select',np.nan,inplace=True)
```

```python
for i in df.columns:
    if (df[i].isnull().mean())*100>5:
        df[i]=df[i].fillna('Unknown')
```

```python
#Dropping rows which still contain null values and also the Prospect ID & Lead Number as they are not required for the model
df=df.dropna()
df.drop(['Prospect ID','Lead Number'],axis=1,inplace=True)
```

```python
df_categorical=df.select_dtypes(include='object')
df_categorical.columns
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
for i in df_categorical:
    df[i]=label_encoder.fit_transform(df[i])
```

```python
df.head()
```

```python
df.corr()
```

```python
#Removing columns which have only single value
df.drop(['Receive More Updates About Our Courses','Update me on Supply Chain Content','Get updates on DM Content','I agree to pay the amount through cheque'],axis=1,inplace=True)
```

```python
df.info()
```

```python

y=df['Converted']
x=df.drop(['Converted'],axis=1)
```

```python
#Splitting data into train and test sets
from sklearn.model_selection import train_test_split
np.random.seed(0)
x_train, x_test,y_train,y_test = train_test_split(x,y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
import statsmodels.api as sm
from sklearn.feature_selection import RFE
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
lm=LogisticRegression()
lm.fit(x_train,y_train)
rfe=RFE(lm,n_features_to_select=20)
rfe=rfe.fit(x_train,y_train)
```

```python

list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python
x_train_rfe=x_train[x_train.columns[rfe.support_]]
x_train_rfe.head()
```

```python
#Build a logistic model


def model_lm(x_train):
 X_train_lm = sm.add_constant(x_train)
 lr = sm.GLM(y_train, X_train_lm,family=sm.families.Binomial()).fit()
 print(lr.summary())
 return lr
    
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
def vif_model(x):
   vif = pd.DataFrame()
   vif['Features'] = x.columns
   vif['VIF'] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
   vif['VIF'] = round(vif['VIF'], 2)
   vif = vif.sort_values(by = "VIF", ascending = False)
   return vif
```

```python
lr=model_lm(x_train_rfe)
```

```python

vif=vif_model(x_train_rfe)
vif
```

```python
#Dropping variables only with p value more than 0.05 and VIF greater than 10
drop_list=['Do Not Call','Do Not Email','Search','Newspaper','Through Recommendations','City']
x_train_rfe=x_train_rfe.drop(drop_list,axis=1)
```

```python
lr=model_lm(x_train_rfe)
```

```python
vif=vif_model(x_train_rfe)
vif
```

# Prediction using the model and comparision of actual vs predicted

```python
y_train_pred = lr.predict(sm.add_constant(x_train_rfe))
```

```python
y_train_pred=y_train_pred.values.reshape(-1)
```

```python
y_train_final=pd.DataFrame({'Actual':y_train.values,'Pred_Probability':y_train_pred})
y_train_final['ID']=y_train.index
y_train_final.head()
```

```python
y_train_final['Predicted']=y_train_final.Pred_Probability.map(lambda x : 1 if x>0.5 else 0)
y_train_final.head()
```

```python
#Confusion matrix
from sklearn import metrics
c_matrix=metrics.confusion_matrix(y_train_final['Actual'],y_train_final['Predicted'])
print(c_matrix)
```

```python
#Accuracy score
print(metrics.accuracy_score(y_train_final['Actual'],y_train_final['Predicted']))
```

```python
#Recall score
print(metrics.recall_score(y_train_final['Actual'],y_train_final['Predicted']))
```

# Final Model fitting on test data and evaluation

```python
# Adding constant variable to test dataframe
X_test_lm = sm.add_constant(x_test)
```

```python
X_test_lm.shape
```

```python
# Creating X_test_lm dataframe by dropping variables from X_test_lm

X_test_lm = X_test_lm.drop(['Do Not Call','Do Not Email','Search','Newspaper','Through Recommendations','City','Magazine',
       'Newspaper Article', 'X Education Forums', 'Digital Advertisement'], axis = 1)
```

```python
# Making predictions using the built model on test data

y_test_pred = lr.predict(X_test_lm)
y_test_pred=y_test_pred.values.reshape(-1)
y_test_final=pd.DataFrame({'Actual_Test':y_test.values,'Pred_Probability_Test':y_test_pred})
y_test_final['ID']=y_test.index
y_test_final.head()
y_test_final['Predicted_Test']=y_test_final.Pred_Probability_Test.map(lambda x : 1 if x>0.5 else 0)
y_test_final.head()
```

```python
#Confusion matrix for Test data
from sklearn import metrics
c_matrix_test=metrics.confusion_matrix(y_test_final['Actual_Test'],y_test_final['Predicted_Test'])
print(c_matrix_test)
```

```python
#Accuracy score
print(metrics.accuracy_score(y_test_final['Actual_Test'],y_test_final['Predicted_Test']))
```

```python
#Recall score
print(metrics.recall_score(y_test_final['Actual_Test'],y_test_final['Predicted_Test']))
```
